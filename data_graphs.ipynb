```python
# Import Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the parquet table and do a quick test
parquet_file_path = 'gza_data/clinical_data/PhonAID_clinical.parquet'
df = pd.read_parquet(parquet_file_path)
print(df.head())
```

```python
# Plotting Histogram
df['jitter'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Histogram Plot')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating a boxplot of shimmer values
axes = df.groupby('location').boxplot(by = 'diagnosis', column =['jitter'], grid = False)
for ax in axes:
    ax.set_ylim(-0.5,10)
    ax.set_ylabel('jitter')
    ax.set_title('')
```

```python
# Bar plot
ax = df['diagnosis'].value_counts().plot(kind='bar', xlabel='Diagnosis', ylabel='Counts', rot=30)
```
